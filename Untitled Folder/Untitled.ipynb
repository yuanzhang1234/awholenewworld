```python
from tkinter.ttk import *
from tkinter import *
sc = Tk()
sc.title("Caro 3x3")
sc.geometry("400x530")

height = 3
width = 7
relief = "raised"
font = ("Arial",16,'bold')
bg = 'skyblue1'
fg = 'green'
count_set = 1
count_x = 0
count_o = 0
check = 'O'
count_turn = 0
draw_check = True
win_check = False
win_lists = [["1","2","3"], ["4","5","6"],["7","8","9"],["1","4","7"],["2","5","8"],["3","6","9"],["1","5","9"],["3","5","7"]]
select_x =[]
select_o =[]

def win_color(no):
    if no == "1":
        bt1.config(bg = "royalblue")
    elif no == "2":
        bt2.config(bg = "royalblue")
    elif no == "3":
        bt3.config(bg = "royalblue")
    elif no == "4":
        bt4.config(bg = "royalblue")
    elif no == "5":
        bt5.config(bg = "royalblue")
    elif no == "6":
        bt6.config(bg = "royalblue")
    elif no == "7":
        bt7.config(bg = "royalblue")
    elif no == "8":
        bt8.config(bg = "royalblue")
    elif no == "9":
        bt9.config(bg = "royalblue")
def winner():
    global draw_check
    global win_check
    global count_x
    global count_o
    for i in range(0,8):
        check_winx = 0
        check_wino = 0
        for j in win_lists[i]:
            if j in select_x:
                check_winx +=1
                if check_winx == 3:
                    lb1.config(text= 'X wins', fg = 'red')
                    draw_check = False
                    win_check = True
                    count_x += 1
                    lb2.config(text = 'X:     ' + str(count_x))
                    lb5.config(text='')
                    lb6.config(text='')
                    for k in win_lists[i]:
                        win_color(k)
                    break
            if j in select_o:
                check_wino +=1
                if check_wino == 3:
                    lb1.config(text= 'O wins', fg = 'darkgreen')
                    draw_check = False
                    win_check = True
                    count_o += 1
                    lb3.config(text = str(count_o) + "     :O")
                    lb5.config(text='')
                    lb6.config(text='')
                    for k in win_lists[i]:
                        win_color(k)
                    break
def drawgame():
    if count_turn == 9 and draw_check == True:
        lb1.config(text= 'Draw', fg = 'royalblue')
        lb5.config(text='')
        lb6.config(text='')
def click_no(bt_no, no):
    global check
    global count_turn
    count_turn +=1
    bt_no["state"] = "disable"    
    if check == 'O':
        check = 'X'
        bt_no.config(text='X', disabledforeground="red")
        lb5.config(text='')
        lb6.config(text='O turn')
        select_x.append(no)     
    else:
        check = 'O'
        bt_no.config(text='O', disabledforeground="darkgreen")
        lb5.config(text='X turn')
        lb6.config(text='')
        select_o.append(no)
    bt_no["state"] = "disable"
    winner()
    drawgame()
    if win_check == True:
        bt1["state"] = "disable"
        bt2["state"] = "disable"
        bt3["state"] = "disable"
        bt4["state"] = "disable"
        bt5["state"] = "disable"
        bt6["state"] = "disable"
        bt7["state"] = "disable"
        bt8["state"] = "disable"
        bt9["state"] = "disable"

def nhan_1():
    click_no(bt1, "1")
def nhan_2():
    click_no(bt2, "2")
def nhan_3():
    click_no(bt3, "3")
def nhan_4():
    click_no(bt4, "4")
def nhan_5():
    click_no(bt5, "5")
def nhan_6():
    click_no(bt6, "6")
def nhan_7():
    click_no(bt7, "7")
def nhan_8():
    click_no(bt8, "8")
def nhan_9():
    click_no(bt9, "9")
def nhan_10():
    bt1.config(text ="", state = "normal", bg = 'skyblue1')
    bt2.config(text ="", state = "normal", bg = 'skyblue1')
    bt3.config(text ="", state = "normal", bg = 'skyblue1')
    bt4.config(text ="", state = "normal", bg = 'skyblue1')
    bt5.config(text ="", state = "normal", bg = 'skyblue1')
    bt6.config(text ="", state = "normal", bg = 'skyblue1')
    bt7.config(text ="", state = "normal", bg = 'skyblue1')
    bt8.config(text ="", state = "normal", bg = 'skyblue1')
    bt9.config(text ="", state = "normal", bg = 'skyblue1')
    lb1.config(text ='')
    global select_x
    global select_o
    global count_turn
    global draw_check
    global win_check
    global count_set
    global check
    select_x =[]
    select_o =[]
    count_turn = 0
    draw_check = True
    win_check = False
    count_set += 1
    lb4.config(text ='SET\n ' + str(count_set))
    if check == 'O':
        lb5.config(text='X turn')
        lb6.config(text='') 
    else:
        lb5.config(text='')
        lb6.config(text='O turn')
    
bt1 = Button(sc, text="", height=height, width=width,relief = relief,font = font,bg = bg,fg = fg, command = nhan_1)
bt1.grid(column=1, row=1,padx = (50,0), pady = (60,0))
bt2 = Button(sc, text="",height=height, width=width,relief = relief,font = font,bg = bg,fg = fg, command = nhan_2)
bt2.grid(column=2, row=1, pady = (60,0))
bt3 = Button(sc, text="", height=height, width=width,relief = relief,font = font,bg = bg,fg = fg, command = nhan_3)
bt3.grid(column=3, row=1, pady = (60,0))
bt4 = Button(sc, text="", height=height, width=width,relief = relief,font = font,bg = bg,fg = fg, command = nhan_4)
bt4.grid(column=1, row=2,padx = (50,0))
bt5 = Button(sc, text="", height=height, width=width,relief = relief,font = font,bg = bg,fg = fg, command = nhan_5)
bt5.grid(column=2, row=2)
bt6 = Button(sc, text="", height=height, width=width,relief = relief,font = font,bg = bg,fg = fg, command = nhan_6)
bt6.grid(column=3, row=2)
bt7 = Button(sc, text="", height=height, width=width,relief = relief,font = font,bg = bg,fg = fg, command = nhan_7)
bt7.grid(column=1, row=3,padx = (50,0))
bt8 = Button(sc, text="", height=height, width=width,relief = relief,font = font,bg = bg,fg = fg, command = nhan_8)
bt8.grid(column=2, row=3)
bt9 = Button(sc, text="", height=height, width=width,relief = relief,font = font,bg = bg,fg = fg, command = nhan_9)
bt9.grid(column=3, row=3)
lb1 = Label(sc,text = '',font = ("Arial",15))
lb1.grid(column = 2, row = 4)
lb5 = Label(sc,text = 'X turn',font = ("Arial",15), fg = 'red')
lb5.grid(column = 1, row = 4)
lb6 = Label(sc,text = '',font = ("Arial",15), fg = 'darkgreen')
lb6.grid(column = 3, row = 4)
lb2 = Label(sc,text = 'X:     -',font = ("Arial",15), fg = 'red', height=2, width=9, relief = "raised")
lb2.grid(column = 1, row = 5, pady = (20,0),padx = (40,0))
lb3 = Label(sc,text = '-     :O',font = ("Arial",15), fg = 'darkgreen', height=2, width=8, relief = "raised")
lb3.grid(column = 3, row = 5, pady = (20,0))
lb4 = Label(sc,text = 'SET\n 1',font = ("Arial",15))
lb4.grid(column = 2, row = 5, pady = (20,0))
bt10 = Button(sc, text="Restart", height=1, width=7, font = ("Arial",15),relief = "raised",bg = "skyblue1",fg = "black", command = nhan_10)
bt10.grid(column = 2, row = 6)
sc.mainloop()
```
